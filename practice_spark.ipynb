```python
one_through_9 = range(1,10)
parallel = sc.parallelize(one_through_9, 3)
def f(iterator): yield sum(iterator)
parallel.mapPartitions(f).collect()
```

```python
print("5.mapPartitionsWithIndex")
```

```python
print("6.GroupByKey")
```

```python
x = sc.parallelize([
    ("USA", 1), ("USA", 2), ("India", 1),
    ("UK", 1), ("India", 4), ("India", 9),
    ("USA", 8), ("USA", 3), ("India", 4),
    ("UK", 6), ("UK", 9), ("UK", 5)], 3)

```

```python
y = x.groupByKey()
```

```python
y.collect()
```

```python
print('Output: ',y.getNumPartitions()) 
```

```python
for t in y.collect():
    print(t[0], [v for v in t[1]])
```

```python
print("#MAP EXAMPLE")
x = sc.parallelize(["spark", "rdd", "example",  "sample", "example"], 2)
y = x.map(lambda x: (x,1))
print(y.collect())
# [('spark', 1), ('rdd', 1), ('example', 1), ('sample', 1), ('example', 1)]

# Another example of making tuple with string and it's length
y = x.map(lambda x: (x,len(x)))
print(y.collect())
# [('spark', 5), ('rdd', 3), ('example', 7), ('sample', 6), ('example', 7)]
```
